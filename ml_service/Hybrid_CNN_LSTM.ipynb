```python
import numpy as np
import os
import tensorflow as tf
from tensorflow.keras import layers, models, callbacks
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tqdm import tqdm
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns

# 1. Chargement des données avec ordre forcé des classes
def load_data_from_npz(base_path):
    X, y = [], []
    classes = ['cassure', 'sain', 'desiquilibre']  # Ordre explicite
    
    for label in classes:
        folder_path = os.path.join(base_path, label)
        files = [f for f in os.listdir(folder_path) if f.endswith(".npz")]
        
        for file in tqdm(files, desc=f"Chargement {label}"):
            with np.load(os.path.join(folder_path, file)) as data:
                if len(data.files) < 9:
                    raise ValueError(f"Le fichier {file} ne contient que {len(data.files)} signaux (9 requis)")
                signals = [data[key] for key in sorted(data.files)[:9]]
                sample = np.stack(signals, axis=0)
                X.append(sample)
                y.append(label)
    
    # CORRECTION : Encodage manuel pour garantir l'ordre
    y_enc = np.array([classes.index(label) for label in y])  # cassure=0, sain=1, desiquilibre=2
    return np.array(X, dtype=np.float32), y_enc, classes  # Retourne aussi les classes

base_path = "/kaggle/input/motorbd-timeshift10/MotorBD_timeshift10"
X, y_enc, classes = load_data_from_npz(base_path)  # Récupère l'ordre des classes
np.save('/kaggle/working/class_order.npy', classes)  # Sauvegarde l'ordre
print(f"Ordre des classes : {classes}")

# 2. Prétraitement (identique)
X = np.transpose(X, (0, 2, 1))  # (n_samples, 50001, 9)
X = (X - np.mean(X, axis=1, keepdims=True)) / (np.std(X, axis=1, keepdims=True) + 1e-8)

# 3. Séparation des données (identique)
X_train, X_test, y_train, y_test = train_test_split(
    X, y_enc, test_size=0.2, stratify=y_enc, random_state=42
)

# 4. Modèle 
def create_model():
    model = models.Sequential([
        layers.Input(shape=(50001, 9)),
        layers.Conv1D(64, 11, activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.MaxPooling1D(4),
        layers.Dropout(0.2),
        layers.Conv1D(128, 7, activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.MaxPooling1D(4),
        layers.Dropout(0.2),
        layers.Conv1D(256, 5, activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.MaxPooling1D(2),
        layers.Dropout(0.3),
        layers.Bidirectional(layers.LSTM(128, return_sequences=True)),
        layers.Bidirectional(layers.LSTM(64)),
        layers.Dense(128, activation='relu'),
        layers.BatchNormalization(),
        layers.Dropout(0.5),
        layers.Dense(3, activation='softmax')  # Sortie correspond à l'ordre des classes
    ])
    return model

model = create_model()
model.compile(
    optimizer=tf.keras.optimizers.Adam(1e-4),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
print("\n=== ARCHITECTURE DU MODÈLE ===")
model.summary()
# 5. Entraînement 
history = model.fit(
    X_train, y_train,
    epochs=8,
    batch_size=16,
    validation_data=(X_test, y_test),
    callbacks=[
        callbacks.EarlyStopping(patience=7, restore_best_weights=True),
        callbacks.ReduceLROnPlateau(factor=0.2, patience=5, min_lr=1e-6)
    ]
)
# 6- Courbes d'apprentissage 
def plot_learning_curves(history):
    plt.figure(figsize=(12, 5))

    # Courbe de la perte (loss)
    plt.subplot(1, 2, 1)
    plt.plot(history.history['loss'], label='Entraînement')
    plt.title('Courbe de la perte')
    plt.xlabel('Épochs')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)

    # Courbe de l'exactitude (accuracy)
    plt.subplot(1, 2, 2)
    plt.plot(history.history['accuracy'], label='Entraînement')
    plt.title('Courbe de l’exactitude')
    plt.xlabel('Épochs')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    plt.show()

plot_learning_curves(history)
# 7. Sauvegarde
model.save('/kaggle/working/cnn_lstm_motor_model_fixedd.h5')
print("Modèle sauvegardé avec l'ordre des classes fixé.")

# 8. Évaluation avec matrice de confusion
def evaluate_model(model, X_test, y_test, classes):
    y_pred = model.predict(X_test)
    y_pred_classes = np.argmax(y_pred, axis=1)

    # Matrice de confusion
    cm = confusion_matrix(y_test, y_pred_classes)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=classes,
                yticklabels=classes)
    plt.xlabel('Prédictions')
    plt.ylabel('Vraies étiquettes')
    plt.title('Matrice de Confusion (Ordre des classes: {})'.format(classes))
    plt.show()

    # Rapport de classification
    print(classification_report(y_test, y_pred_classes, target_names=classes))
    
    test_loss, test_acc = model.evaluate(X_test, y_test, verbose=0)
    print(f"\n Précision finale : {test_acc:.2%}")

# Appel avec l'ordre des classes
evaluate_model(model, X_test, y_test, classes)
```

```python
import numpy as np
import scipy.io
import os
from tensorflow.keras.models import load_model
import matplotlib.pyplot as plt
from tqdm import tqdm
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Configuration
TEST_FOLDER = "/kaggle/input/test-motor/Test_MotorBD_DeepLearning"
MODEL_PATH = '/kaggle/working/cnn_lstm_motor_model_fixedd.h5'
CLASS_ORDER_PATH = '/kaggle/working/class_order.npy'
OUTPUT_FOLDER = '/kaggle/working/converted_test_filess'
REQUIRED_SIGNALS = ['i1', 'i2', 'i3', 'v1', 'v2', 'v3', 'vn', 'w_m', 'vibrad']

# Charger le modèle et les classes
model = load_model(MODEL_PATH)
classes = np.load(CLASS_ORDER_PATH)
os.makedirs(OUTPUT_FOLDER, exist_ok=True)

def convert_mat_folder(root_folder, output_folder):
    """Convertit tous les fichiers .mat en .npz en extrayant les signaux requis"""
    converted_files = []
    for root, dirs, files in os.walk(root_folder):
        for file in tqdm(files, desc=f"Conversion {os.path.basename(root)}"):
            if file.endswith(".mat"):
                mat_path = os.path.join(root, file)
                try:
                    mat_data = scipy.io.loadmat(mat_path, struct_as_record=False, squeeze_me=True)
                    main_key = next(k for k in mat_data if not k.startswith("__"))
                    signals = {}
                    for entry in mat_data[main_key].Y:
                        try:
                            signal_name = entry.Name.split('/')[-2].replace('"', '')
                            if signal_name in REQUIRED_SIGNALS:
                                signals[signal_name] = entry.Data
                        except AttributeError:
                            continue

                    missing = [sig for sig in REQUIRED_SIGNALS if sig not in signals]
                    if missing:
                        raise ValueError(f"Missing signals: {missing}")

                    if len(next(iter(signals.values()))) != 50001:
                        raise ValueError("Signal length != 50001")

                    processed = {name: (data - np.mean(data)) / (np.std(data) + 1e-8)
                                 for name, data in signals.items()}

                    parent_folder = os.path.basename(root)
                    save_name = f"{parent_folder}_{os.path.splitext(file)[0]}.npz"
                    save_path = os.path.join(output_folder, save_name)
                    np.savez_compressed(save_path, **processed)
                    converted_files.append((parent_folder, save_path))

                except Exception as e:
                    print(f"\n⚠ Erreur dans {file} ({os.path.basename(root)}): {str(e)}")
    return converted_files

def predict_batch(model, npz_paths):
    """Effectue les prédictions sur tous les fichiers convertis"""
    results = []
    for etat_reel, npz_path in tqdm(npz_paths, desc="Prédiction"):
        try:
            with np.load(npz_path) as data:
                if set(data.files) != set(REQUIRED_SIGNALS):
                    raise ValueError("Signaux incomplets après conversion")
                sample = np.stack([data[key] for key in sorted(REQUIRED_SIGNALS)], axis=-1)[np.newaxis, ...]
                sample = (sample - np.mean(sample, axis=1, keepdims=True)) / (np.std(sample, axis=1, keepdims=True) + 1e-8)
                pred = model.predict(sample, verbose=0)[0]
                pred_class = classes[np.argmax(pred)]
                confidence = np.max(pred)
                results.append({
                    'etat_reel': etat_reel,
                    'fichier': os.path.basename(npz_path),
                    'prediction': pred_class,
                    'confiance': float(confidence),
                    'probabilities': {cls: float(prob) for cls, prob in zip(classes, pred)}
                })
        except Exception as e:
            print(f"\n❌ Erreur prédiction {npz_path}: {str(e)}")
    return results

# Exécution principale
print("=== DÉBUT DE LA CONVERSION ===")
converted_files = convert_mat_folder(TEST_FOLDER, OUTPUT_FOLDER)
print(f"\n✅ {len(converted_files)} fichiers convertis avec succès")

print("\n=== DÉBUT DES PRÉDICTIONS ===")
predictions = predict_batch(model, converted_files)

print("\n=== RÉSULTATS FINAUX ===")
print("Etat Réel    | Fichier               | Prédiction | Confiance | Probabilités")
print("-"*90)

y_true = []
y_pred = []

for res in predictions:
    probs_str = ", ".join([f"{k}:{v:.2%}" for k,v in res['probabilities'].items()])
    print(f"{res['etat_reel'][:12].ljust(12)} | {res['fichier'][:40].ljust(40)} | "
          f"{res['prediction'].ljust(10)} | {res['confiance']:.2%} | {probs_str}")
    y_true.append(res['etat_reel'])
    y_pred.append(res['prediction'])

# Matrice de confusion
plt.figure(figsize=(10, 8))
cm = confusion_matrix(y_true, y_pred, labels=classes)
cmd = ConfusionMatrixDisplay(cm, display_labels=classes)
cmd.plot(xticks_rotation=45, cmap='Blues')
plt.title("Matrice de Confusion: Etat Réel vs Prédit")
plt.tight_layout()
plt.show()

# Visualisation des confiances
plt.figure(figsize=(10, 6))
for i, cls in enumerate(classes):
    confidences = [p['confiance'] for p in predictions if p['prediction'] == cls]
    plt.scatter([i]*len(confidences), confidences, alpha=0.5, label=cls)
plt.xticks(range(len(classes)), classes)
plt.title("Distribution des Confiances par Classe Prédite")
plt.ylabel("Confiance")
plt.ylim(0, 1.1)
plt.grid(True)
plt.legend()
plt.show()
```
